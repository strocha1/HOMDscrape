# Module 1
## Extracting sequence identifiers from an xml BLAST file

```python
from bs4 import BeautifulSoup
import re
```

```python
# reading content from xml file
file = open(r"C:\Users\strocha\Downloads\sequenceserver-xml_report (2).xml")
contents = file.read()
```

```python
# parsing the file to extract species names
soup = BeautifulSoup(contents, 'xml')
names = soup.find_all('Hit_def')
```

```python
# display species names content from xml tags
for data in names:
    print(data.get_text())
```

## Module 2 
### Extracting amino acid sequences 

```python
# parsing the file to extract sequences
soup = BeautifulSoup(contents, 'xml')
sequences = soup.find_all('Hsp_qseq')
```

```python
# display sequences from xml tags
for data in sequences:
    print("\n" + data.get_text())
```

```python
seq = [] # Empty list for sequences
for data in sequences:
    seq.append(data.get_text()) # saving text from xml file to seq list
```

## Module 3 
### Extracting names from sequence IDs and formatting for FASTA use

```python
na = [] # generating an empty list for names 
for data in names:
    na.append("\n" + data.get_text()) # appending the items to a list while adding /n for later splitting
    print(data.get_text()) # displaying the content
```

```python
def extract_first_bracket_text_from_list(text_list):
    # Define the regex pattern to find text within the first set of brackets from na list
    pattern = re.compile(r'\[(.*?)\]')
    
    # List to store the extracted text
    extracted_texts = []
    
    # Iterate over each line in the list
    for line in text_list:
        # Find all matches of the pattern in the line
        matches = pattern.findall(line)
        if matches:
            # Append the first match to the extracted_texts list
            extracted_texts.append(matches[0])
    
    return extracted_texts

# Extract the text
extracted_names = extract_first_bracket_text_from_list(na)

# Print the results
for text in extracted_names:
    print(text)
```

## Module 4 
### Combining the species names and amino acid sequences into a single document

```python
final1=list(zip(extracted_texts,seq)) #combining the extracted_names list and the seq list 
print(final1)
```

```python
f = open('finalfile.txt', 'w') # Opening a new text file for output
for t in final1:
    line = '\n'.join(str(x) for x in t)
    f.write("\n>" + line)
f.close() # saving the final list as a text document and closing the document
```
